```python
from matplotlib import pyplot as plt
from astropy.table import Table
import numpy as np
```

```python
tau_params = Table.read('rand_skewers_z45_ovt_tau.fits', hdu=1)
tau_skewers = Table.read('rand_skewers_z45_ovt_tau.fits', hdu=2)
```

```python
tau_skewers.columns
```

```python
oden = tau_skewers['ODEN'].flatten()
temp = tau_skewers['T'].flatten()
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(121)
plt.hist(oden, bins=np.arange(0., 10, 0.2), histtype='step');
plt.xlabel(r'overdensity ($\rho/\bar{\rho}$)', fontsize=15)

plt.subplot(122)
plt.hist(oden, bins=np.arange(0., 100, 0.2), histtype='step', cumulative=True, density=True);
plt.xlabel(r'overdensity ($\rho/\bar{\rho}$)', fontsize=15)
plt.ylabel('CDF', fontsize=15)
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(121)
plt.hist(temp, bins=np.arange(1000, 3e4, 50), histtype='step');
plt.xlabel('T(K)', fontsize=15)

plt.subplot(122)
plt.hist(temp, bins=np.arange(1000, 3e4, 50), histtype='step', cumulative=True, density=True);
plt.xlabel('T(K)', fontsize=15)
plt.ylabel('CDF', fontsize=15)
```
